```python
%matplotlib inline
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('lena.jpg')
b,g,r = cv2.split(img)
img2 = cv2.merge([r,g,b])
plt.imshow(img2) # expect true color
```

```python
img3 = img2[:256]
plt.imshow(img3)
```

```python
img4 = img2[:, 0:256]
plt.imshow(img4)
```

```python
img5 = img2[256:512, 256:512]
plt.imshow(img5)
```
